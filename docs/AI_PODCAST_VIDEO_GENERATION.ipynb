## 3. Extract & Clean Scripts for AI Tools

```python
# Extract the podcast narration (without production notes)
# Remove markdown headers and metadata

def extract_narration(script_text, start_marker='## OPENING', end_marker='## END'):
    """Extract clean narration from markdown script"""
    lines = script_text.split('\n')
    narration = []
    capture = False
    
    for line in lines:
        if '**Voice-Over:**' in line or '**Host:**' in line:
            # Extract the voiceover content
            content = line.split('**')[-1].strip()
            if content:
                narration.append(content)
        elif '**[' in line and ']**' in line:
            # Skip stage directions but capture them as comments
            pass
    
    return '\n\n'.join(narration)

# Extract podcast narration
podcast_narration = extract_narration(podcast_script)

# Save clean podcast script for NotebookLM
podcast_clean_path = GENERATED_DIR / 'podcast' / 'PODCAST_CLEAN.txt'
with open(podcast_clean_path, 'w') as f:
    f.write(podcast_narration)

print(f"✓ Extracted podcast narration ({len(podcast_narration)} characters)")
print(f"✓ Saved to: {podcast_clean_path}")
print(f"\n--- First 300 characters ---\n{podcast_narration[:300]}...")
```
